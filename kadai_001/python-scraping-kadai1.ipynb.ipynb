```python
import requests
from bs4 import BeautifulSoup

url = "https://news.yahoo.co.jp/articles/2e28db4de1de728ebd04ff54adf386b21c2a7df8"

res = requests.get(url)
res.raise_for_status()

soup = BeautifulSoup(res.text, "html.parser")

# タイトル（articlesページはこれで安定）
print("■ タイトル")
title = soup.find("h1")
print(title.get_text(strip=True))

# 本文（articlesページの安定構造）
print("\n■ 本文")
body = soup.find("div", {"class": "article_body"}) or soup.find("article")

if body is None:
    raise RuntimeError("記事本文が取得できません")

for p in body.find_all("p"):
    text = p.get_text(strip=True)
    if text:
        print(text)
```
